# Relacionando dados de diferentes dataframes

## PIB, Desmatamento e Gases do efeito estufa do setor Agropecuário

Para comparar os dados de PIB e desmatamento é necessário plotar os gráficos, porém, para isso, é necessário comparar apenas os mesmos municípios. Para isso, retirou-se os municípios do Data Frame "df_pib_municipios" que não estão no dataframe "df_desmatamento".

```python
#Para comparar os dataframes df_desmatamento e df_pib_municipios, é necessário q eles tenham
# a mesma quantidade de linhas. Para isso, apenas as linhas com municipios em comum.

lista_municipios = df_desmatamento['id_municipio'].unique()
for index, row in df_pib_municipios.iterrows():
    if row[0] not in lista_municipios:
        df_pib_municipios.drop(index, inplace = True, axis = 0)
```

```python
print(len(df_pib_municipios['id_municipio'].unique()),len(df_desmatamento['id_municipio'].unique()))
```

```python
plt.figure(figsize = (16,10))
plt.title('Evolução VA setor agropecuário')
sns.lineplot(x='ano',y='VA_agropecuaria', data = df_pib_municipios)
```

```python
plt.figure(figsize = (16,10))
plt.title('Evolução do PIB nos municipios brasileiros com o passar dos anos')
sns.lineplot(x='ano',y='PIB', data = df_pib_municipios)
```

```python
plt.figure(figsize = (16,10))
plt.title('Evolução do Desmatamento')
sns.lineplot(x='ano',y='desmatado', data = df_desmatamento)
```

```python
plt.figure(figsize = (16,6))
plt.title('Emissão de Gases do Efeito estufa do setor Agropecuario')
sns.lineplot(x=df_estufa_uf[df_estufa_uf['ano']>=2000]['ano'],
             y=df_estufa_uf[df_estufa_uf['nivel_1']=='Agropecuária']['emissao'],
             data = df_estufa_uf)
```
