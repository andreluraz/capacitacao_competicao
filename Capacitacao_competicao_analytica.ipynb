## PIB, Desmatamento e Gases do efeito estufa do setor Agropecuário

Para comparar os dados de PIB e desmatamento é necessário plotar os gráficos, porém, para isso, é necessário comparar apenas os mesmos municípios. Para isso, retirou-se os municípios do Data Frame "df_pib_municipios" que não estão no dataframe "df_desmatamento".

Observando os dois graficos, percebe-se que à medida que o PIB dos municípios aumenta, o desmatamento também aumenta, indicando uma possível correlação entre esses dados. Isso pode ser explicado pois quanto mais as atividades se desenvolvem, mais espaço é necessário para expandí-las, aumentando o desmatamento. Entretando, em países em que a população tem maior acesso à informação, essa reta tende a diminuir, pois a mesma percebe o grande impacto negativo que isso pode causar. 

No caso dos municípios do brasil, percebe-se um aumento na taxa de desmatamento, indicando uma falta de informação da população desses municípios.
